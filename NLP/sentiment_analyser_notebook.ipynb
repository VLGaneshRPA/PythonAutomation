```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
nltk.download('twitter_samples')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
text1 = 'Hey, what a beautiful day! How amazing it is!'
```

```python
if analyzer.polarity_scores(text1)['compound'] > 0:
    print('Positive Text')
else:
    print('Negative Text')
```

```python
len(nltk.corpus.twitter_samples.strings())
```

```python
#number 1045 is the random tweet number
tweet1 = nltk.corpus.twitter_samples.strings()[2000]
tweet1
```

```python
analyzer.polarity_scores(tweet1)
```

```python
if analyzer.polarity_scores(tweet1)['compound'] > 0:
    print('Positive Text')
else:
    print('Negative Text')
```
